```python
import pandas as pd

def fasta_to_dataframe(fasta_file):
    headers = []
    sequences = []
    with open(fasta_file, 'r') as f:
        current_sequence = []
        current_header = None
        for line in f:
            line = line.strip()
            if line.startswith('>'):  # Header line
                if current_header is not None:
                    # Save the previous sequence
                    headers.append(current_header)
                    sequences.append(''.join(current_sequence))
                current_header = line[1:]  # Remove '>'
                current_sequence = []
            else:
                current_sequence.append(line)
        # Add the last sequence
        if current_header is not None:
            headers.append(current_header)
            sequences.append(''.join(current_sequence))
    # Create a DataFrame
    return pd.DataFrame({'Header': headers, 'Sequence': sequences})

fasta_file = "../data/DB.COX1.fna"
full_dataset = fasta_to_dataframe(fasta_file)

full_dataset[['BOLD Metadata', 'Taxonomy']] = full_dataset['Header'].str.split(';', n=1, expand=True)
full_dataset[['BOLD ID', 'Read ID', 'Country']] = full_dataset['BOLD Metadata'].str.split('|', n=2, expand=True)
full_dataset['BOLD ID'] = full_dataset['BOLD ID'].str.removeprefix('BOLD:')

tax_columns = ['Phylum', 'Class', 'Order', 'Family', 'Genus', 'Species']
full_dataset[tax_columns] = full_dataset['Taxonomy'].str.extract(
    r'p:([^,]+),c:([^,]+),o:([^,]+),f:([^,]+),g:([^,]+),s:([^;]+)'
)

full_dataset = full_dataset.drop(columns=['Header', 'BOLD Metadata', 'Taxonomy'])
print(full_dataset.head())
```

```python
my_result = pd.read_csv("../data/primer-finder-result-0.csv", sep=",")
my_result['region_length'] = my_result['read'].str.len()
```

```python
my_result['f_length'] = my_result['f_match'].str.len()
my_result['b_length'] = my_result['b_match'].str.len()
```

```python
my_result = my_result.merge(
    full_dataset[["Read ID", "Sequence"]],
    on="Read ID",
    how="left"
)
```

```python
my_result['f_match_2'] = my_result.apply(
    lambda row: row['Sequence'][row['f_index']:row['f_index'] + len(row['f_match'])]
    if pd.notna(row['Sequence']) and pd.notna(row['f_index']) and pd.notna(row['f_match'])
    else None,
    axis=1
)
no_nan = my_result[~my_result['f_match'].isna()]
no_skips = no_nan[~no_nan['f_match'].str.contains('-')]
no_skips[no_skips['f_match_2'] != no_skips['f_match']]
```

```python
critical_seqs = my_result[my_result['f_index'] < 0]
no_skips = critical_seqs[~critical_seqs['f_match'].str.contains('-')]
no_skips.loc[:,'f_index'] = 0
no_skips['f_match_3'] = no_skips.apply(
    lambda row: row['Sequence'][row['f_index']:row['f_index'] + len(row['f_match'])]
    if pd.notna(row['Sequence']) and pd.notna(row['f_index']) and pd.notna(row['f_match'])
    else None,
    axis=1
)
no_skips
```

```python
my_result['b_match_2'] = my_result.apply(
    lambda row: row['Sequence'][row['b_index']:row['b_index'] + len(row['b_match'])]
    if pd.notna(row['Sequence']) and pd.notna(row['b_index']) and pd.notna(row['b_match'])
    else None,
    axis=1
)
no_nan = my_result[~my_result['b_match'].isna()]
no_skips = no_nan[~no_nan['b_match'].str.contains('-')]
no_skips[no_skips['b_match_2'] != no_skips['b_match']]
```
