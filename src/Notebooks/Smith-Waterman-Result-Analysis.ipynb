```python
import pandas as pd

df = pd.read_csv("../../data/primer-finder-multi4.csv", sep=";")
print(df.head())
```

```python
# foreach order
    # create statistic over all f_primer and b_primer
    # search top 5 families
        # also create statistic

forward_family_counts = df.groupby(['Order', 'Family', 'f_match']).size().reset_index(name='Count')
back_family_counts = df.groupby(['Order', 'Family', 'b_match']).size().reset_index(name='Count')
forward_family_counts.sort_values(by=['Order', 'Family', 'Count'], ascending=[True, True, False])
back_family_counts.sort_values(by=['Order', 'Family', 'Count'], ascending=[True, True, False])
```

```python
forward_order_counts = df.groupby(['Order', 'f_match']).size().reset_index(name='Count')
back_order_counts = df.groupby(['Order', 'b_match']).size().reset_index(name='Count')
forward_order_counts.sort_values(by=['Order', 'Count'], ascending=[True, False])
back_order_counts.sort_values(by=['Order', 'Count'], ascending=[True, False])
```

```python
with pd.ExcelWriter('../../data/primer-counts.xlsx', engine='openpyxl') as writer:
    forward_family_counts.sort_values(by=['Order', 'Family', 'Count'], ascending=[True, True, False]).to_excel(writer, sheet_name="forward_family", index=False)
    forward_order_counts.sort_values(by=['Order', 'Count'], ascending=[True, False]).to_excel(writer, sheet_name="forward_order",  index=False)
    back_family_counts.sort_values(by=['Order', 'Family', 'Count'], ascending=[True, True, False]).to_excel(writer, sheet_name="back_family",  index=False)
    back_order_counts.sort_values(by=['Order', 'Count'], ascending=[True, False]).to_excel(writer, sheet_name="back_order",  index=False)
```
